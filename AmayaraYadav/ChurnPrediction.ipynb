####  REMOVING ZERO VARIANCE VARIABLE

 ##### zero variance variable->these are variables in a dataset that have the same value for all observations. This means there is no variability in the data for that variable.

### OUTLIER TREATMENT

#### these are the data points that deviates from the rest of the data.
#### it is advisable drop them them but in some cases they play and important role like disease detection
#### there are various ways to detect outliers-
##### 1. boxplot using IQR
##### 2. flooring and capping
##### 3. 3 sigma approach


### +/-3 sigma approach
##### is a statistical method used to identify outliers in a dataset based on the distribution of its values
##### the data points that fall below mean-3*(sigma) or above mean+3*(sigma) are outliers, where mean and sigma are the average value and standard deviation of a particular column.
##### outliers are defined as values that fall outside the range of μ±3σ.

```python
import pandas as pd
def standardize_and_filter( data):
    standardized_df = data.copy()
    for column in data.columns:
        mean =data[column].mean()
        std =data[column].std()
        standardized_df[column] = (data[column] - mean) / std
    filtered_df = standardized_df[(standardized_df >= -3).all(axis=1) & (standardized_df <= 3).all(axis=1)]
    return filtered_df

filtered_df = standardize_and_filter(data)
print(filtered_df.head())
filtered_df.shape
```

## REMOVING MULTICOLLINEARITY 
it arises when variables are correlated, making it challenging to discern their separate effects on the target variable.
#### Variance Inflation Factor(VIF)-> it is an technique to detect multicollinearity##### it can be calculated by: 1/(1-R*R) where, R-squared is the coefficient of determination in linear regression. 
##### Its value lies between 0 and 1.
##### greater the VIF<<greater the correlation.
 

```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor

data_no_corr = data_no_corr.copy() 

def calculate_vif(data):
    return pd.DataFrame({
        'feature': data.columns,
        'VIF': [variance_inflation_factor(data.values, i) for i in range(data.shape[1])]
    })

threshold = 5
while True:
    vif_data = calculate_vif(data_no_corr)
    max_vif = vif_data['VIF'].max()
    if max_vif > threshold:
        feature_to_remove = vif_data.loc[vif_data['VIF'].idxmax(), 'feature']
        print(f"Removing feature '{feature_to_remove}' with VIF: {max_vif}")
        data_no_corr = data_no_corr.drop(columns=[feature_to_remove])
    else:
        break

# Remaining features
remaining_features = data_no_corr.columns

print("Remaining features after removing high VIF features:")
print(remaining_features)
```

### CROSS VALIDATION BEFORE HYPERPARAMETER TUNING
   #### Hyperparameter tuning -> process of selecting the optimal values for a machine learning model’s hyperparameters
#### The goal of hyperparameter tuning is to find the values that lead to the best performance on a given task.


```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from imblearn.over_sampling import SMOTE


X = data_no_corr.drop('target', axis=1)  # Features
y = data_no_corr['target']  # Target variable

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=45)

smote = SMOTE(random_state=45)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

# Initialize StandardScaler
scaler = StandardScaler()

# Fit the scaler on the training data and transform both training and test sets
X_train_res_scaled = scaler.fit_transform(X_train_res)
X_test_scaled = scaler.transform(X_test)

# Initialize SGDClassifier
sgd_model = SGDClassifier(random_state=45)

# Perform cross-validation
cv_scores = cross_val_score(sgd_model, X_train_res_scaled, y_train_res, cv=5, scoring='accuracy')
print(f'Cross-validation scores: {cv_scores}')
print(f'Mean CV Accuracy: {cv_scores.mean()}')

# Train the model on the training set
sgd_model.fit(X_train_res_scaled, y_train_res)

# Predict on the test set
y_pred = sgd_model.predict(X_test_scaled)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

```

## HYPERPARAMETER TUNING USING GRID SEARCH CROSS VALIDATION
We fit the model using all possible combinations after creating a grid of potential discrete hyperparameter values. We log each set’s model performance and then choose the combination that produces the best results. This approach is called GridSearchCV, because it searches for the best set of hyperparameters from a grid of hyperparameters values. 

```python
import pandas as pd
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
```

```python

# Define the parameter grid to search through
param_grid = {
    'n_estimators': [20, 30,40 ],   # no. of trees
    'max_depth': [10, 20, 30,],       # maximum depth
    'min_samples_split': [2, 5, 10, ],    # min. no. of sample required to split an internal code
    'min_samples_leaf': [2, 4, 6, ],          # min. no. of  sample required to be a leaf node
}

model = RandomForestClassifier(random_state=45)

grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='accuracy')

grid_search.fit(X_train, y_train)

# Print the best parameters and best score found by GridSearchCV
print("Best Parameters:", grid_search.best_params_)
print("Best CV Score:", grid_search.best_score_)

# Get the best estimator from grid search
best_model = grid_search.best_estimator_

# Predict on the test set with the best model
y_pred = best_model.predict(X_test)

```

#### CROSS VALIDATION AFTER HYPERPARAMETER TUNING

#### Cross validation-> technique used in machine learning to evaluate the performance of a model on unseen data.
    It involves dividing the available data into multiple folds or subsets, using one of these folds as a validation set, and training the model on the remaining folds. This process is repeated multiple times, each time using a different fold as the validation set. Finally, the results from each validation step are averaged to produce a more robust estimate of the model’s performance.
   ###### helps to ensure that the model selected for deployment is robust and generalizes well to new data.

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
import matplotlib.pyplot as plt

# Assuming data_no_corr is already loaded and preprocessed
X = data_no_corr.drop(columns='target')
y = data_no_corr['target']

X = X.fillna(X.mean())

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=50)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Define the reduced parameter grid for hyperparameter tuning
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

# Initialize the Random Forest model
rf_model = RandomForestClassifier(random_state=50)

# Use GridSearchCV to find the best parameters
grid_search_rf = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=3, scoring='accuracy', n_jobs=-1, verbose=1)
grid_search_rf.fit(X_train, y_train)

# Print the best parameters and best score found by GridSearchCV
print("Best Parameters:", grid_search_rf.best_params_)
print("Best CV Score:", grid_search_rf.best_score_)

# Get the best estimator from grid search
best_rf_model = grid_search_rf.best_estimator_

# Predict on the test set with the best model
y_pred = best_rf_model.predict(X_test)

# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

conf_matrix = confusion_matrix(y_test, y_pred)

# Print evaluation metrics
print(f'Accuracy: {accuracy:.2f}')
print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'F1 Score: {f1:.2f}')

print('Confusion Matrix:')
print(conf_matrix)
```

##### Root Node: It is the topmost node in the tree, which represents the complete dataset. It is the starting point of the decision-making process.
##### 
Internal Node: A node that symbolizes a choice regarding an input feature.
.##### 
Leaf/Terminal Node: A node without any child nodes that indicates a class label or a numerical value
##### .
Parent Node: The node that divides into one or more child nodelit.
